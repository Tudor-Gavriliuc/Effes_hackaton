```python
from sklearn.ensemble import IsolationForest
```

```python
df1423.info()
```

```python
def get_IsolationForest(df, target):
    df_IsolationForest = df1423.copy()
    model=IsolationForest()
    # Apply algorithm to data
    df_IsolationForest['outlier']=pd.Series(model.fit_predict(df_IsolationForest.values))
    # Remove rows where there is -1 in created column
    df_IsolationForest=df_IsolationForest[df_IsolationForest.outlier!=-1]
    # Drop created column
    df_IsolationForest.drop(['outlier'], axis=1, inplace=True)
    return df_IsolationForest

df_IsolationForest = get_IsolationForest(df1423,'nr_unit')
df_IsolationForest
```

```python
from sklearn.ensemble  import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

X = df_IsolationForest.drop(['nr_unit'], axis = 1)
Y = df_IsolationForest['nr_unit']

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.33, random_state = 12)
regr = RandomForestRegressor()
regr.fit(X_train,y_train)
f44 = regr.predict(X_test)
print(r2_score(y_test,f44))
print(mean_absolute_error(y_test,f44))
print(mean_squared_error(y_test,f44))
```

```python
from sklearn.neural_network import MLPRegressor

regr = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
f34 = regr.predict(X_test)
print(r2_score(y_test,f34))
```

```python
from sklearn.neighbors import KNeighborsRegressor

neigh = KNeighborsRegressor()
neigh.fit(X_train, y_train)
f33 = neigh.predict(X_test)
print(r2_score(y_test,f33))
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
 
regr.fit(X_train, y_train)
f44 = regr.predict(X_test)
print(r2_score(f44, y_test))
```

```python
from sklearn.decomposition import PCA


def PCA_2(df,n_comp,target):
    X = df.drop(target, axis = 1)
    Y = df[target]
    pca = PCA(n_components=n_comp)
    pca.fit(X)
    pca_df = pca.transform(X)
    pca_df = pd.DataFrame(pca_df)
    pca_df[target] = Y
    return pca_df

pca_df = PCA_2(df_IsolationForest,20,'nr_unit')
```

```python
pca_df['nr_unit'] = df_IsolationForest['nr_unit'].values
pca_df
```

```python
X = pca_df.drop(['nr_unit'], axis = 1)
Y = pca_df['nr_unit']

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.33, random_state = 12)
regr = RandomForestRegressor()
regr.fit(X_train,y_train)
f44 = regr.predict(X_test)
print(r2_score(y_test,f44))
```
