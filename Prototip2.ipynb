```python
from sklearn.ensemble import IsolationForest
def get_IsolationForest(df, target):
    df_IsolationForest = filtered_df1.copy()
    model=IsolationForest()
    # Apply algorithm to data
    df_IsolationForest['outlier']=pd.Series(model.fit_predict(df_IsolationForest.values))
    # Remove rows where there is -1 in created column
    df_IsolationForest=df_IsolationForest[df_IsolationForest.outlier!=-1]
    # Drop created column
    df_IsolationForest.drop(['outlier'], axis=1, inplace=True)
    return df_IsolationForest

df_IsolationForest = get_IsolationForest(filtered_df1,'nr_unit')
df_IsolationForest
```

```python
from sklearn.decomposition import PCA


def PCA_2(df,n_comp,target):
    X = df.drop(target, axis = 1)
    Y = df[target]
    pca = PCA(n_components=n_comp)
    pca.fit(X)
    pca_df = pca.transform(X)
    pca_df = pd.DataFrame(pca_df)
    pca_df[target] = Y
    return pca_df

pca_df = PCA_2(filtered_df1,20,'nr_unit')
```

```python
from kydavra import PValueSelector

def get_PValue(df,target):
    pvalue_selector = PValueSelector()
    selected_columns = pvalue_selector.select(df,target)
    return selected_columns
    
selected_columns = get_PValue(filtered_df1,'nr_unit')
selected_columns
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
def apply_model(df,model):
    X = df.drop('nr_unit', axis = 1)
    Y = df['nr_unit']

    X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.25, random_state = 12)
    tuned_parameters = {
    'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson']
    }
    clf = GridSearchCV(RandomForestRegressor(), tuned_parameters,scoring='recall')
    clf.fit(X_train, y_train)
    output69 = clf.predict(X_test)
    f1 = mean_squared_error(y_test,output69)
    f2 = mean_absolute_error(y_test,output69)
    f3 = r2_score(y_test,output69)
    print(f1,' ',f2,' ',f3)
    
def fun1(model):
    print("Simple df")
    apply_model(filtered_df1,model)
    print("PCA dataframe")
    apply_model(pca_df,model)
```
